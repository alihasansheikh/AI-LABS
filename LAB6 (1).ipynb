```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Book1.csv')
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
plt.show()
```

```python
new_df = df.drop('price',axis='columns')
new_df
```

```python
price =df.price
price
```

```python
reg = LinearRegression()
reg.fit(new_df,price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('pizza.csv')
```

```python
%matplotlib inline
plt.xlabel('size')
plt.ylabel('price')
plt.scatter(df.pizza,df.price,color='red',marker='+')
plt.show()
```

```python
new_df = df.drop('price',axis='columns')
new_df
```

```python
price =df.price
price
```

```python
reg = LinearRegression()
reg.fit(new_df,price)
```

```python
predict=[[24],[30],[32]]
arr=np.array(reg.predict(predict))
newarr = arr.astype('i')
print("new size :" ,end="")
print(predict)
print("new price :" ,end="")
print(newarr)
```

```python
reg.coef_
```

```python
reg.intercept_
```
